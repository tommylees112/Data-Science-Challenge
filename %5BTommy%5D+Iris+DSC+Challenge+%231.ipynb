# Load data and convert to pandas DataFrame

```python
from sklearn import datasets
iris = datasets.load_iris()
# dir(iris) - explore the features in the dataset
# iris['target_names'] - what are the species names?
```

```python
iris_df = pd.DataFrame(data = np.c_[iris['data'],iris['target']], columns=iris['feature_names']+['target'])
iris_df
```

np.c_ is the numpy concatenate function which is used to concat iris['data'] and iris['target'] arrays

for pandas column argument: concat iris['feature_names'] list and string list (in this case one string); you can make this anything you'd like.. 

the original dataset would probably call this ['Species']

# Append the Species names as a column on the end of the dataframe

```python
species = [iris.target_names[num] for num in iris.target]
species = np.where(iris.target==0, iris.target_names[0],
          np.where(iris.target==1, iris.target_names[1], iris.target_names[2]))
```

```python
iris_df['species'] = species
iris_df
```

```python
def iris_species(x):
    if x['target'] == 0.0:
        return 'setosa'
    elif x['target'] == 1.0:
        return 'versicolor'
    elif x['target'] == 2.0:
        return 'virginica'
    else:
        return 'undefined'

iris['species'] = (iris_species,iris)
```
