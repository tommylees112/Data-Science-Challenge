# Load data and convert to pandas DataFrame

http://stackoverflow.com/questions/38105539/how-to-convert-a-scikit-learn-dataset-to-a-pandas-dataset

https://github.com/justmarkham/DAT3/blob/master/code/08_iris_solutions.py

```python
#get the data
from sklearn import datasets
iris = datasets.load_iris()

# dir(iris) - explore the features in the dataset
# iris['target_names'] - what are the species names?
```

```python
#load into a panda dataframe, using the 'data' & 'target' values & the column headers (feature names & target)
iris_df = pd.DataFrame(data = np.c_[iris['data'],iris['target']], columns=iris['feature_names']+['target'])

# clean up variable names (label as 'features')
features = [name[:-5].replace(' ', '_') for name in iris.feature_names]

#rename colum headings & remove 'target' (could be done at the start) 
iris_df = pd.DataFrame(iris.data, columns=features)
iris_df
```

np.c_ is the numpy concatenate function which is used to concat iris['data'] and iris['target'] arrays

for pandas column argument: concat iris['feature_names'] list and string list (in this case one string); you can make this anything you'd like.. 

the original dataset would probably call this ['Species']

# Append the Species names as a column on the end of the dataframe

```python
#put the species name into a list ???

species = [iris.target_names[num] for num in iris.target]
species = np.where(iris.target==0, iris.target_names[0],
          np.where(iris.target==1, iris.target_names[1], iris.target_names[2]))
```

```python
# put the species names into the 'iris_df' dataframe
iris_df['species'] = species
iris_df
```

```python
multi_plot2 = sns.PairGrid(iris_df, hue="species")
multi_plot2.map_diag(plt.hist)
multi_plot2.map_offdiag(plt.scatter)
multi_plot2.add_legend();
```

### Try and create a new column of 'species' based on conditional on 'target' column

### OR Do it with a function

http://stackoverflow.com/questions/27041724/using-conditional-to-generate-new-column-in-pandas-dataframe 

```python
def iris_species(x):
    if x['target'] == 0.0:
        return 'setosa'
    elif x['target'] == 1.0:
        return 'versicolor'
    elif x['target'] == 2.0:
        return 'virginica'
    else:
        return 'undefined'

iris['species'] = (iris_species,iris)
```
